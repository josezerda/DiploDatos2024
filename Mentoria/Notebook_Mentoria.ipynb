```python
import pandas as pd
import matplotlib.pyplot as plt


pd.options.display.max_columns=None
```

## Vamos a cargar los datos en formato csv

**Cargamos las mediciones diarias por mes**

```python
df_diario_enero = pd.read_csv("Clou_diario_diplo/Enero_FINAL.csv", sep=";")
df_diario_febrero = pd.read_csv("Clou_diario_diplo/Febrero_FINAL.csv", sep=";")
df_diario_marzo = pd.read_csv("Clou_diario_diplo/Marzo_FINAL.csv", sep=";")
df_diario_abril = pd.read_csv("Clou_diario_diplo/Abril_FINAL.csv", sep=";")
df_diario_mayo = pd.read_csv("Clou_diario_diplo/Mayo_FINAL.csv", sep=";")
df_diario_junio = pd.read_csv("Clou_diario_diplo/Junio_FINAL.csv", sep=";")
df_diario_julio = pd.read_csv("Clou_diario_diplo/Julio_FINAL.csv", sep=";")
df_diario_agosto = pd.read_csv("Clou_diario_diplo/Agosto_FINAL.csv", sep=";")
df_diario_septiembre = pd.read_csv("Clou_diario_diplo/Septiembre_FINAL.csv", sep=";")
df_diario_octubre = pd.read_csv("Clou_diario_diplo/Octubre_FINAL.csv", sep=";")
df_diario_noviembre = pd.read_csv("Clou_diario_diplo/Noviembre_FINAL.csv", sep=";")
df_diario_diciembre = pd.read_csv("Clou_diario_diplo/Diciembre_FINAL.csv", sep=";")

```

**Combinamos los dataframes por mes en uno solo**

```python
df_diario_combinado = pd.concat([df_diario_enero, df_diario_febrero, df_diario_marzo, df_diario_abril, df_diario_mayo, df_diario_junio, df_diario_julio, df_diario_agosto, df_diario_septiembre, df_diario_octubre, df_diario_noviembre, df_diario_diciembre], ignore_index=True)
```

```python
df_diario_combinado
```

**Cargamos el Dataframe de datos de ubicacion**

```python
df_datos_gis = pd.read_csv("Clou_diario_diplo/QGIS_final.csv", sep=";")
```

```python
df_datos_gis['grupo_tarifario'].value_counts()
```

```python
df_datos_gis.head(3)
```

**Agregamos al dataframe diario combinado, los datos referentes a la ubicacion**

```python
new_columns = ['suministro', 'contrato', 'fecha_inicio', 'sea', 'distribuidor', 'subestacion', 'distrito_descripcion', 'grupo_lectura', 'ruta_lectura', 'orden_lectura', 'grupo_tarifario', 'codigo_tarifa', 'tarifa'] 
```

```python
for col in new_columns:
    df_diario_combinado[col] = pd.NA
```

```python
df_diario_combinado
```

**A continuacion vamos a completar el dataframe combinado de todos los meses con los datos de georeferencia**

```python
for index, row in df_datos_gis.iterrows():
    df_diario_combinado.loc[df_diario_combinado['DEVICE'] == row['DEVICE'], ['suministro', 'contrato', 'fecha_inicio', 'sea', 'distribuidor', 'subestacion', 'distrito_descripcion', 'grupo_lectura', 'ruta_lectura', 'orden_lectura', 'grupo_tarifario', 'codigo_tarifa', 'tarifa']] = pd.DataFrame({'suministro': row['suministro'], 'contrato': row['contrato'], 'fecha_inicio': row['fecha_inicio'], 'sea': row['sea'] , 'distribuidor': row['distribuidor'], 'subestacion': row['subestacion'], 'distrito_descripcion': row['distrito_descripcion'], 'grupo_lectura': row['grupo_lectura'], 'ruta_lectura': row['ruta_lectura'], 'orden_lectura': row['orden_lectura'], 'grupo_tarifario': row['orden_lectura'], 'codigo_tarifa': row['codigo_tarifa'], 'tarifa': row['tarifa']}, index=df_diario_combinado.index[df_diario_combinado['DEVICE'] == row['DEVICE']])
```

### El dataframe df_diario_combinado ahora tiene los datos de localizacion y tipo de consumidor

```python
df_diario_combinado
```

## Analizamos el dataframe de ubicacion

### Cuantos usuarios por tipo de tarifa

```python
conteos = df_datos_gis['tarifa'].value_counts()

# Crear el gráfico de barras
conteos.plot(kind='bar')

# Personalizar el gráfico
plt.title('Usuarios por tipo de tarifa')
plt.xlabel('Tarifas')
plt.ylabel('Usuarios')

# Mostrar el gráfico
plt.show()
```

## Usuarios por distribuidor

```python
conteos = df_datos_gis['distribuidor'].value_counts()

# Crear el gráfico de barras
conteos.plot(kind='bar')

# Personalizar el gráfico
plt.title('Usuarios por Distribuidor')
plt.xlabel('Distribuidor')
plt.ylabel('Usuarios')

# Mostrar el gráfico
plt.show()
```
